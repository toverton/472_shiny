### **Methods** 

### *Recurrent Neural Networks (RNNs)*

Recurrent Neural Networks (RNNs) are a class of neural networks specially designed to handle sequence data, where the order and context of the elements are important. Unlike traditional neural networks which process inputs independently, the looping mechanism of RNNs enables information to be carried over from one input step to the next. This makes them ideal for tasks that involve temporal sequences such as speech recognition, language modeling, and time series prediction .

How RNNs Work? 

RNN processes sequences by repeating through the input data points and keeping up a state that contains information learned from previous data points. Here is a breakdown of the process:

1- Input Layer: Each element of the sequence is input sequentially into the RNN. This could be a word in a sentence, a frame in a video, or a time point in a stock price series.

2- Hidden Layer: The RNN has a hidden state that stores information about the sequence processed so far. At each step in the sequence, the hidden state is updated based on two inputs: the new input data and the previous hidden state. This update is often a simple function like a tanh, sigmoid or rule.

3- Output Layer: After processing each element, the RNN can output something or it can continue processing until it reaches the end of the sequence, at which point it might output a final result.

Unlike traditional neural networks, which learn separate parameters for each input feature, RNNs share parameters across different parts of the model. This not only reduces the total number of parameters in the model but also improves its efficiency and scalability when processing long sequences (Buhl, 2023).

Limitations:

The vanishing gradient problem presents a significant obstacle in training deep neural networks, such as Recurrent Neural Networks (RNNs). This issue arises when the gradientskey indicators that guide the adjustment of neural weights during trainingdiminish in size as they move back through the network's layers. This reduction can severely impede an RNN's capacity to grasp long-distance relationships within the data, often resulting in sluggish or inefficient training outcomes. This problem is especially acute in scenarios requiring the recall or transmission of information across extended periods, which compromises the networks ability to identify crucial patterns. To tackle this challenge, various strategies have been developed. Methods such as optimizing the initial weights, applying batch normalization, clipping gradients, incorporating skip connections, and adjusting learning rates are critical in maintaining steady gradient flows and averting their premature decline. Amid this spectrum of approaches, two standout solutions take center stage: LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) (Buhl, 2023).


### *Long Short-Term Memory Networks (LSTM)*


Long Short-Term Memory Networks, or LSTMs, are a type of deep, sequential neural network designed to retain information over extended periods. They are a specialized variant of Recurrent Neural Networks (RNNs) specifically developed to address the vanishing gradient issue common in standard RNNs. Created by Hochreiter and Schmidhuber, LSTMs effectively overcome the challenges posed by traditional RNNs and other machine learning algorithms, making them particularly adept at processing sequences where context and memory are crucial (Saxena, 2024).


How LSTM Work?



In the first part of an LSTM cycle, the network decides whether to retain or discard information from the previous timestamp, determining its relevance to current computations. The second part involves the cell attempting to absorb new information from the input it receives. Finally, in the third part, the cell updates and transmits this refined information to the next timestamp. This entire process, which spans these three stages, constitutes a single timestep within the LSTM operation.

In an LSTM unit, the three crucial components known as gates are responsible for managing the flow of information through the memory cell, or LSTM cell. These gates include the Forget gate, which decides what information to discard from the cell state. The Input gate which determines what new information to add to the cell state. The Output gate which controls what information from the cell state to pass to the output. These gates and the LSTM cell function similarly to a layer of neurons in a traditional feedforward neural network, where each neuron possesses a hidden layer and maintains a current state. Like a basic RNN, an LSTM also maintains a hidden state, where H(t1) represents the hidden state from the previous timestamp and is the hidden state at the current timestamp. Additionally, LSTMs feature a cell state, denoted as  C(t1) for the previous timestamp and C(t) for the current one. the hidden state is often referred to as the short-term memory, reflecting the network's recent past information, while the cell state is considered the long-term memory, holding information over longer sequences. These techniques allow LSTMs to retain a more complex and extended memory of past data, crucial for tasks requiring the analysis of patterns over time(Saxena, 2024).

### *Gated Recurrent Units (GRU)*

GRUs are very similar to LSTM (Long Short Term Memory) GRU controls the information flow using gates, just like LSTM does. Compared to LSTM, they are relatively new. This explains why their architecture is simpler and they provide some improvement over LSTM. There are two gates in a GRU as opposed to three gates in an LSTM cell. The first gate is the reset gate and the other one is the update gate. The Reset gate is care for the short-term memory of the network (the hidden state (Ht)). Update Gate (long-term memory) Similarly, we have an Update gate for long-term memory.



It takes the hidden state Ht-1 from the previous timestamp t-1 and an input Xt at each timestamp t. It then outputs a new hidden state, Ht, which is passed to the following timestamp once more. The interesting thing about GRU is that, unlike LSTM, it does not have a separate cell state (Ct). There is just one hidden state for it (Ht). GRUs train more quickly because of their less complex architecture(Saxena, 2023). 


How GRU Work?

GRUs manage information flow using two gates: the reset gate and the update gate. The reset gate determines how much previous information to retain, enabling the network to forget unnecessary details, which helps in maintaining only relevant information from the past. The update gate controls how much new information is used to update the hidden state. This dual-gate mechanism allows GRUs to handle both recent and distant information effectively, making them superior for tasks involving long-term dependencies, where traditional RNNs might fail due to the vanishing gradient problem. By selectively updating their hidden states, GRUs can capture complex patterns in sequential data, learning from both short-term and long-term (Stewart, 2024).


### Table that captures the key differences between Recurrent Neural Networks (RNNs), Long Short-Term Memory Networks (LSTMs), and Gated Recurrent Units (GRUs):


| Feature                      | RNNs                            | LSTMs                                                     | GRUs                                          |
|------------------------------|---------------------------------|-----------------------------------------------------------|-----------------------------------------------|
| **Core Idea**                | Handles sequence data through looping mechanisms. | Specifically designed to avoid the vanishing gradient problem by using gates. | Simpler version of LSTMs with fewer gates.   |
| **Architecture**             | Simple loop with hidden states. | Three gates (Forget, Input, Output) and cell states to manage information flow. | Two gates (Reset, Update) with no separate cell state. |
| **Parameter Complexity**     | Less complex, fewer parameters. | High complexity due to additional gates and cell state.    | Less complex than LSTMs due to fewer gates.  |
| **Memory Handling**          | Struggles with long-term dependencies. | Efficient in handling long-term dependencies.              | Effectively balances short and long-term dependencies. |
| **Typical Use Cases**        | Short sequence prediction tasks. | Complex sequence modeling, especially where long-term memory is crucial. | Tasks requiring efficient computation and good handling of long-term dependencies without the full complexity of LSTMs. |
| **Training Performance**     | Faster training but prone to vanishing gradient issues. | Slower training due to complexity but excellent at capturing long-term relationships. | Faster training than LSTMs due to simpler architecture. |
| **Optimal Use Scenario**     | Suitable for less complex scenarios where model simplicity and speed are prioritized. | Preferred for highly complex tasks with significant long-term dependencies. | Ideal when a balance is needed between performance and computational demands. |

```python
selected_df['date'] = pd.to_datetime(selected_df['date'], errors='coerce')



selected_df['year'] = selected_df['date'].dt.year

selected_df.loc[:, 'date'] = pd.to_datetime(selected_df['date'])
selected_df.loc[:, 'year'] = selected_df['date'].dt.year

daily_trends = selected_df.groupby('date').agg(
    total_fatalities=pd.NamedAgg(column='n_killed', aggfunc='sum'),
    total_injured=pd.NamedAgg(column='n_injured', aggfunc='sum'),
    number_of_incidents=pd.NamedAgg(column='date', aggfunc='size')
).reset_index()

# the reason that i remove 2013 because there is an issue with data that we saw it in the EDA
daily_trends_14_18 = daily_trends[
    (daily_trends['date'] >= pd.to_datetime("2014-01-01")) & 
    (daily_trends['date'] <= pd.to_datetime("2017-12-31"))
]

print(daily_trends_14_18.head(), daily_trends_14_18.tail())
```

```python
training = daily_trends_14_18[
    (daily_trends_14_18['date'] >= pd.to_datetime("2014-01-01")) & 
    (daily_trends_14_18['date'] < pd.to_datetime("2017-01-01"))
]

validation  = daily_trends_14_18[
    (daily_trends_14_18['date'] >= pd.to_datetime("2016-09-01")) & 
    (daily_trends_14_18['date'] < pd.to_datetime("2017-01-01"))
]

# Create the testing set with dates from January 1, 2017 to December 31, 2017
testing = daily_trends_14_18[
    (daily_trends_14_18['date'] >= pd.to_datetime("2017-01-01")) & 
    (daily_trends_14_18['date'] <= pd.to_datetime("2017-12-31"))
]

# Display the first few rows to verify
print("Training set:")
print(training.head())
print("validation set:")
print(validation .head())
print("\nTesting set:")
print(testing.head())
```

```python
# Scale data

scaler = MinMaxScaler()
scaled_train=scaler.fit_transform(training[["total_fatalities"]])
scaled_test=scaler.transform(testing[["total_fatalities"]])
scaled_validation=scaler.transform(validation[["total_fatalities"]])
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
n_input=36
n_features=1
train_generator=TimeseriesGenerator(scaled_train,scaled_train, n_input, batch_size=1)
validation_generator = TimeseriesGenerator(scaled_validation, scaled_validation, n_input, batch_size=1)
```

### **Results**

**Long Short Term Memory (LSTM) Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam

model = Sequential([
    # Input layer with specified input shape
    LSTM(150, activation='relu', input_shape=(n_input,n_features), return_sequences=True),
    Dropout(0.2),  # Regularization with dropout

    LSTM(100 , return_sequences=True),
    Dropout(0.2),  # Regularization with dropout
    
    # Additional LSTM Layer
    LSTM(100, return_sequences=True),
    Dropout(0.2),  # Regularization with dropout

    # Bidirectional LSTM Layer
    Bidirectional(LSTM(75, return_sequences=False)),
    Dropout(0.2),  # Regularization with dropout
    
    # Output layer with linear activation
    Dense(1, activation='linear')
])
model.compile(
    loss='MSE',
    optimizer=Adam()
)


model.summary()
```

```python

history = model.fit(
    train_generator,  
    epochs=150,
    validation_data=validation_generator,
    batch_size=50)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

# Plot the loss and validation loss
plt.figure(figsize=(10, 6))
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss',color='red')
plt.title('Training and Validation Loss over Epochs by (LSTM)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
```

```python
y_pred = []

first_batch = scaled_train[-n_input:]
current_batch = first_batch.reshape(1, n_input, n_features)

for i in range(len(scaled_test)):
    batch = current_batch
    pred = model.predict(batch)[0]
    y_pred.append(pred)
    current_batch = np.append(current_batch[:,1:, :], [[pred]], axis = 1)


y_pred
```

```python
test_predictions_array = np.array(y_pred).reshape(-1, n_features)
actual_predictions = scaler.inverse_transform(test_predictions_array)[:, 0] 
```

```python
from sklearn.metrics import mean_squared_error
actual_test_values = scaled_test
actual_test_values = scaler.inverse_transform(actual_test_values)[:, 0]
rmse = np.sqrt(mean_squared_error(actual_test_values, actual_predictions))
print("Root Mean Square Error (RMSE):", rmse)
```

### **Model Discussion**

When comparing the performance of different types of recurrent neural networks (RNNs), such as Simple RNN, LSTM, and GRU, on the same task and architecture, we can observe differences in their effectiveness and efficiency:


- **Simple RNN Model:**

Parameters: Approximately 94,551, which equates to about 369.34 KB of memory usage.

Time: Approximately 75 minutes.

RMSE: The RMSE produced by the model was 21.3458. This is the highest of the three models, indicating that although Simple RNNs can perform sequence prediction tasks, the vanishing gradient problem may cause them to falter when dealing with longer-term dependencies or more complicated patterns.


- **LSTM (Long Short-Term Memory) Model:**

Parameters: Approximately higher at 377,751,  which equates to about to 1.44 MB of memory usage.

Time: Approximately 118 minutes.

RMSE: 10.6795, much lower. This marked improvement in prediction accuracy over the Simple RNN is due to the LSTM's ability to maintain long-term dependencies within the data. LSTMs are designed to overcome the limitations of Simple RNNs by incorporating memory cells that help retain important information over longer sequences.

- **GRU (Gated Recurrent Unit) Model:**

Parameters: Approximately 284,851, which equates to about 1.09 MB of memory usage.

Time: Approximately 131 minutes.

RMSE: RMSE of 12.271, which is better than the Simple RNN but slightly worse than the LSTM. GRUs simplify the LSTM architecture by using two gates (update and reset) and typically require fewer parameters than LSTMs. They provide a good balance between computational efficiency and the ability to capture dependencies in sequence data.


- **Effectiveness:** The LSTM model proves to be the most effective in terms of lowering the RMSE, which indicates better predictive accuracy. This is because LSTMs are particularly strong at capturing long-term dependencies in data sequences, a critical advantage in many predictive modeling scenarios.


- **Efficiency and Simplicity:** Although LSTMs perform best, GRUs offer a more streamlined architecture that makes them computationally less intensive while still capturing complex patterns in the data quite effectively. This makes GRUs a compelling choice when computational resources are a concern or when the additional complexity of LSTMs does not translate into proportional performance gains.

- **Choosing the Right Model:** The choice between Simple RNN, LSTM, and GRU would depend on the specific requirements of the task, including the complexity of the sequence patterns, the importance of capturing long-term dependencies, and computational resource constraints. LSTMs are ideal for highly complex tasks with significant long-term dependencies like the forecast of the deaths in our data, while GRUs provide a middle ground, and Simple RNNs might be suitable for less complex scenarios where model simplicity and faster training times are prioritized.
